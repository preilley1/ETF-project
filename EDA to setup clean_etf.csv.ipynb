### Import  Pandas to create data frames

```python
import pandas as pd
```

### Pull in csv of all the prices of the etfs that were downloaded

```python
#vgit = pd.read_csv("./VGIT interm bond.csv") 
iyr = pd.read_csv("./IYR real estate.txt")
eem = pd.read_csv("./EEM Emerging markets.txt")
agg = pd.read_csv("./AGG US bond.txt")
bsv = pd.read_csv("./BSV short bond.csv")  # Need to get more data
emb = pd.read_csv("./EMB emerging markets bond.txt")
iau = pd.read_csv("./IAU gold trust.txt")
iefa = pd.read_csv("./IEFA dev markets.txt")
ifeu = pd.read_csv("./IFEU Euro real estate.txt")
ijh = pd.read_csv("./IJH Mid cap.txt")
slv = pd.read_csv("./SLV silver.txt")
ijr = pd.read_csv("./IJR Small-cap core.txt")
ivv = pd.read_csv("./ishares Core S&P 500 IVV.txt")
```

### Reverse the order of some of the data.  Rename the column names

```python
iyr = iyr[["Date","Close"]].rename(columns ={"Close":"iyr_close"}).iloc[::-1]
eem = eem[["Date","Close"]].rename(columns ={"Close":"eem_close"}).iloc[::-1]
agg = agg[["Date","Close"]].rename(columns ={"Close":"agg_close"}).iloc[::-1]
bsv = bsv[["Date","Close"]].rename(columns ={"Close":"bsv_close"}) 
emb = emb[["Date","Close"]].rename(columns ={"Close":"emb_close"}).iloc[::-1]
iau = iau[["Date","Close"]].rename(columns ={"Close":"iau_close"}).iloc[::-1]
iefa= iefa[["Date","Close"]].rename(columns ={"Close":"iefa_close"}).iloc[::-1]
ifeu = ifeu[["Date","Close"]].rename(columns ={"Close":"ifeu_close"}) # data start in 2012
ijh = ijh[["Date","Close"]].rename(columns ={"Close":"ejh_close"}).iloc[::-1]
slv = slv[["Date","Close"]].rename(columns ={"Close":"slv_close"}).iloc[::-1]
ijr = ijr[["Date","Close"]].rename(columns ={"Close":"ijr_close"}).iloc[::-1]
ivv = ivv[["Date","Close"]].rename(columns ={"Close":"ivv_close"}).iloc[::-1]
#vgit = vgit[["Date","Close"]].rename(columns ={"Close":"vgit_close"}).iloc[::-1]
```

```python
# Merge the data frames
etf = iyr.merge(eem).merge(agg).merge(emb).merge(bsv).merge(emb).merge(iau).merge(ifeu).merge(ijh).merge(slv).merge(ijr).merge(ivv)
```

```python
etf.head()
```

```python
# Lower case all the columns names - this is really just the datae column at this point
etf.columns = map(str.lower, etf.columns)
```

```python
etf['date'] = pd.to_datetime(etf['date']) # creating datetime dtype for date 
```

```python
etf.set_index('date', inplace = True) # setting date to "index" - take a look a next cell
```

```python
etf['ivv_close_shft']= etf['ivv_close'].shift() # creating a new column to shift the close price up 1
```

```python
etf.head()
```

```python
# Dropping one row with a NA value above
etf.dropna(inplace=True)
```
